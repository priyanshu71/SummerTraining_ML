## Making changes to Data

```python
# Adding new column
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.sort_values('Total' , ascending=0)
```

```python
# Dropping the Columns
df = df.drop(columns=['Total'])
df.head()
```

```python
# adding columns alternative way

df['Total'] = df.iloc[: , 4:10].sum(axis=1)  #axis=0 vertically axis=1 Horizontally
df.head(3)
```

```python
cols = list(df.columns.values) 
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head()
```

## Saving our Data (Exporting into the Desired format)

## Filtering Data

```python
df.loc[(df['Type 1'] == 'Fire') & (df['Total'] > 600)] # use & instead of and
```

```python
df.loc[(df['Type 1'] == 'Fire') | (df['Type 2'] == 'Fighting')]
```

```python
# we can also make new dataframe of filtered one
new_df = df.loc[(df['Type 1'] == 'Fire') & (df['Total'] > 600)]
new_df
#If required use, new_df.to_csv('Filtered.csv') to save it
```

### Note in above dataframe, indexes aren't ordered

### To Reset Index

```python
new_df = new_df.reset_index()
new_df
```

```python
# This is storing old indices so to get rid of that
new_df = df.loc[(df['Type 1'] == 'Fire') & (df['Total'] > 600)]
new_df = new_df.reset_index(drop=True)
new_df
```

```python
# Suppose if we want to filter out all Pokemon name that contains word "Mega" in them
df.loc[df['Name'].str.contains('Mega')]
```

```python
# If we put tilde (~) then it will eliminate all name that contains "Mega"
df.loc[~df['Name'].str.contains('Mega')]
```

## Regex Filtering

```python
import re
```

```python
df.loc[df['Type 1'].str.contains('Fire|grass', flags=re.I , regex=True)]
#flags=re.I is to ignorecase otherwise we have to take care of Capitalizations
```

```python
# getting pokemon name that contains 'pi' in their name eg pikachu etc.
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

## Conditional Changes

```python
#suppose if we wanted to change Fire to Flame in Type 1 column
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flame'
df
```

```python
# Lets change all Grass Pokemon to Legendary
df.loc[df['Type 1'] == 'Grass', 'Legendary'] = True
df
```

```python
# we can modify several parameters at a time
df.loc[df['Type 1'] == 'Grass', ['HP' , 'Generation']] = [100 , 'XYZ']
df
```
