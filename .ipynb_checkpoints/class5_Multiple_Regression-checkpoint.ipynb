# Multicollinearity in Linear Regression

```python
import pandas as pd
```

```python

import statsmodels.api as sm #This library consists of Different Regression Model
df_adv = pd.read_csv('Advertising.csv', index_col=0)
X = df_adv[['TV', 'Radio', 'Newspaper']]
y = df_adv['Sales']
df_adv.head()

# sales = b0*1 + b1*tv + b2*radio + b3*newspaper
```

```python
X = sm.add_constant(X)
```

```python
X
```

```python
#fit a OLS model with intercept on TV and Radio

model = sm.OLS(y, X).fit()
model.summary()
```

```python
import matplotlib.pyplot as plt
X.iloc[:,1:].corr()
#.corr() is used to evaluate correlation between the variables
```

```python
df_salary = pd.read_csv('Salary_Data.csv')
df_salary.head()
```

```python
X = df_salary[['YearsExperience', 'Age']]
y = df_salary['Salary']
# salary =  b0*1 + b1*YearsExperienece + b2*Age
```

```python
X = sm.add_constant(X)
X
```

```python
model = sm.OLS(y, X).fit()
model.summary()
```

```python
import matplotlib.pyplot as plt
X.iloc[:,1:].corr()
```

Since both are highly correlated now we have to emit one which has higher p value. Here Age = 0.165 has high p-value so it will be omitted. 

```python
X = df_salary[['YearsExperience']]
X = sm.add_constant(X)
X
```

```python
model = sm.OLS(y, X).fit()
model.summary()
```
